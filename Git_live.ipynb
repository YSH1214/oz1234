- ignore 파일 생성해 버전(commit)으로 만들고 싶지 않은 파일 관리하기
    1. .gitignore 파일 생성
    2. .gitignore 파일 안에 버전(commit)으로 만들고 싶지 않은 파일명 기재

- 버전을 만들이기 위한 add와 commit 명령어

    front.yaml과 back.yaml 파일 생성 후 진행<br>
    Untracked : 관리하고 있지 않는 파일<br>
    Staged : 관리하고 있는 파일, commit 직전에 파일들이 모여있는 곳

    1. Untracked 상태인 파일을 Staged에 올리는 명령어 add
        - git add 파일명  -> 원하는 파일 하나를 등록할 수 있음
        - git add .     -> Untracked 상태인 파일을 모두 Staged에 한번에 등록

    2. Staged에 등록된 파일을 하나의 버전으로 만드는 명렁어 commit
        - git commit -m "커밋 메시지" -> 하나의 버전을 만들었다
        - git commit
            - vim 모드 화면 노출
            - vim 모드에서 키보드의 i 를 눌러 쓰기 모드로 변환
            - 최상단에 commit 메시지 작성 후 키보드의 esc 버튼 누르기
            - shift + ; 눌러 명령어 입력모드로 변환
            - wq 입력 후 엔터를 눌러 저장후 나가기
    
- git의 등록된 commit(버전) 이력 확인하기
    1. git log 명령어를 이용해 이전 commit 이력을 모두 확인하기
        - git log
        - git log --oneline
